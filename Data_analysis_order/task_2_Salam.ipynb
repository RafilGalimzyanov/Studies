```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Способ 1

```python
from mpl_toolkits.mplot3d import Axes3D


x = np.arange (-10, 10, 0.1)
y = np.arange (-10, 10, 0.1)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = np.cos (xgrid) * np.sin (ygrid) / (xgrid * ygrid)

fig = plt.figure()
axes = Axes3D(fig)
axes.plot_surface(xgrid, ygrid, zgrid)
plt.show()
```

## Способ 2

```python
from mpl_toolkits.mplot3d import Axes3D


x = np.arange (-10, 10, 0.1)
y = np.arange (-10, 10, 0.1)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = np.cos (xgrid) * np.sin (ygrid) / (xgrid * ygrid)

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(xgrid, ygrid, zgrid, rstride=4, cstride=4, alpha=0.25)
ax.view_init(30, 45)

ax = fig.add_subplot(1,2,2, projection='3d')
ax.plot_surface(xgrid, ygrid, zgrid, rstride=4, cstride=4, alpha=0.25)
ax.view_init(70, 30)

fig.tight_layout()
```

## Способ 3

```python
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')
```

```python
from datetime import date
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates
# Даты, которые будут отложены по оси X
xdata = [date (2010, 5, 25),
 date (2010, 7, 5),
 date (2010, 12, 1),
 date (2011, 3, 17),
 date (2011, 8, 2),
 date (2011, 11, 13),
 date (2012, 3, 15),
 date (2012, 4, 8),
 date (2012, 12, 21)
 ]
# Данные, которые будут отложены по оси Y
ydata = [0.4, 0.3, 0.5, 0.4, 0.6, 0.3, 0.2, 0.1, 0.0]
# Преобразуем даты в числовой формат
xdata_float = mpl.dates.date2num (xdata)
# Вызовем subplot явно, чтобы получить экземпляр класса AxesSubplot,
# из которого будем иметь доступ к осям
axes = plt.subplot(1, 1, 1)
# Пусть в качестве меток по оси X выводится только год
axes.xaxis.set_major_formatter (mpl.dates.DateFormatter("%Y"))
# Отобразим данные
plt.plot_date (xdata_float, ydata, fmt="b-")
plt.grid()
plt.show()
```
