```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib notebook


def freq_range(n):
    '''
    Функция freq_range принимает в аргументы значение эксперимента, определяет номер попадаемого диапазона и 
    возвращает его
    '''
    if 4.6 < n <= 4.7:
        return 1
    elif 4.7 < n <= 4.8:
        return 2
    elif 4.8 < n <= 4.9:
        return 3
    elif 4.9 < n <= 5.0:
        return 4
    elif 5.0 < n <= 5.1:
        return 5
    elif 5.1 < n <= 5.2:
        return 6
    elif 5.2 < n <= 5.3:
        return 7
    elif 5.3 < n <= 5.4:
        return 8
    
def input_name():
    '''
    Функция input_name запускает считывание фамилии и имени с клавиатуры, 
    возвращает кортеж из букв, содержащие ФИ, отсортированные по алфавиту
    '''
    fi = input('Введите фамилию и имя через пробел: ')
    fi = fi.replace(' ', '')

    individ_let = set([i.upper() for i in fi])
    individ_let = tuple(sorted(individ_let))
    
    return individ_let

def toFixed(numObj, digits=0):
    '''
    Фиксированное количество знаков после запятой
    '''
    return f"{numObj:.{digits}f}"
```

```python
individ_let = input_name()

print(individ_let)

ex = pd.read_excel('data_100.xlsx')

my_data = ex.query(f"Алфавит == {individ_let} ")
my_data.index = np.arange(1, len(my_data)+1)
index = [freq_range(i) for i in my_data['Результат отдельного измерения, ti, с']]

my_data['Окончательный номер'] = np.arange(1, len(my_data)+1)
my_data['Номер ячейки гистограммы'] = index

mean_data = np.mean(my_data['Результат отдельного измерения, ti, с'])

variation = [mean_data - i for i in my_data['Результат отдельного измерения, ti, с']]
variation_sq = [i**2 for i in variation]

my_data['Δt = tср - ti, с'] = variation
my_data['(Δti)^2, с^2'] = variation_sq

summ_delta = sum(my_data['Δt = tср - ti, с'])
summ_delta_sq = sum(my_data['(Δti)^2, с^2'])

my_data
```

```python
print(f'Среднее значение: {mean_data},  \nСуммарное отклонение: {summ_delta},\n'
      f'Суммарный квадрат отклонения: {summ_delta_sq}')
```

```python
### Сохранение данных в excel 

writer = pd.ExcelWriter('RESULT_LR100.xlsx')
my_data.to_excel(writer)
writer.save()
```

```python
### Рисунок 1. График зависимости результата наблюдения от номера измерения

y = my_data['Результат отдельного измерения, ti, с']
x = [i for i in range(len(y))]

plt.scatter(x, y)
plt.grid()
plt.xlabel('Номер измерения')
plt.ylabel('Результат измерения')
plt.title('Рисунок 1. \nГрафик зависимости результата наблюдения от номера измерения')
```

```python
### Рисунок 2. Гистограмма экспериментальных значений и кривая закона распределения

gist_elem = {1, 2, 3, 4, 5, 6, 7, 8}
counts = my_data['Номер ячейки гистограммы']
y = []

for elem in gist_elem:
    y.append(list(counts).count(elem))

groups = [f'{toFixed(4.6+0.1*i,1)}-{toFixed(4.7+0.1*i,1)}' for i in range(len(y))]
plt.bar(groups, y, width=0.95)
plt.grid()
plt.title('Рисунок 2. \n Гистограмма экспериментальных значений и кривая закона распределения')
```
