# Задача на определения муфт на кривой электромагнитного отклика

```python
import csv
import matplotlib.pyplot as plt
import statistics
from scipy.signal import find_peaks
```

```python
data = []

def data_read():
    # Читаем данные из файла и записываем в лист data ---------------------------------------------------------
    with open('2.csv', 'r') as file:
        reader = csv.reader(file) #Чтение файла (итерируемый объект)
        for row in reader: #Проходимся по строкам данных 
            data.append(float(row[0]))
        #end_for
    #end_with
    return data
#end_def
#
# *****************************************************************************

data = data_read()
x = [i for i in range(len(data))] # Задаем координаты по оси "OХ"

plt.figure(figsize=(30, 10))
plt.plot(x, data)

'''
    В задании было сказано, что уровень сигнала пропорционален толщине. Но судя по данным, в местах соединения труб муфтами,
интенсивность сигнала порядка 4/3 от сигнала самой трубы.
    Для поиска величины сигнала от трубы (фактически шума) использовалась медиана от данных трубы умноженная на 1.3
'''
peaks = find_peaks(data, height=statistics.median(data)*1.3, distance=1)
min_y = min(data)

'''
    Вертикальные линии рисовались методом vlines(), цикл создавался для того, чтобы у каждой линии был свой предел сверху.
'''
for elem in peaks[0]:
    plt.vlines(x=int(elem), ymin=min_y, ymax=data[elem], colors='red', lw=1.5, label='vline_multiple - full height')
#end_for

plt.grid()
plt.show()
```

```python
# Координаты и интенсивность пиков 

for elem in peaks[0]:
    print(f'Координата: {elem},  Интенсивность: {data[elem]}')
```
